```python
row_count = duckdb.sql(f"""
        SELECT COUNT(*) FROM '{file_path}/*.parquet'
    """)

print(row_count)
```

```python
columns = ['pickup_datetime', 'dropoff_datetime', 'request_datetime', 'on_scene_datetime', 'trip_miles', 'trip_time', 'base_passenger_fare', 'congestion_surcharge']

for col in columns:
    required = duckdb.sql(f"""
        SELECT COUNT(*) AS {col} FROM '{file_path}/*.parquet'
        WHERE {col} IS NULL
    """)

    print(required)
```

```python
check_time = duckdb.sql(f"""
    SELECT COUNT(*) FROM '{file_path}/*.parquet'
    WHERE pickup_datetime > dropoff_datetime;
""")

print(check_time)
```

```python
check_miles = duckdb.sql(f"""
    SELECT COUNT(*) FROM '{file_path}/*.parquet'
    WHERE trip_miles < 0;
""")

print(check_miles)
```

```python
check_trip_time = duckdb.sql(f"""
    SELECT COUNT(*) FROM '{file_path}/*.parquet'
    WHERE trip_time < 0;
""")

print(check_trip_time)
```
