```python
import duckdb
import matplotlib.pyplot as plt

# laptop
file_path = "C:/Users/idavo/.cache/kagglehub/datasets/jeffsinsel/nyc-fhvhv-data/versions/4"
```

```python
average_2020 = duckdb.sql(f"""
    SELECT 
        strftime(pickup_datetime, '%Y-%m-%d') AS day,
        AVG(base_passenger_fare) AS avg_fare
    FROM "nyc_data.parquet"
    WHERE strftime(pickup_datetime, '%Y') = '2020'
    GROUP BY day
    ORDER BY day
""")

print(average_2020)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

avg_df_2020 = average_2020.df()

# Convert 'day' column to datetime
avg_df_2020['day'] = pd.to_datetime(avg_df_2020['day'])

plt.figure(figsize=(12, 6))
plt.plot(avg_df_2020['day'], avg_df_2020['avg_fare'], label="Avg Fare")

# Customize plot
plt.xlabel("Month")
plt.ylabel("Average Fare ($)")
plt.title("Average Fare by Day in 2020")

# Generate tick locations (start of each month)
tick_locations = pd.date_range(start=avg_df_2020['day'].min(), 
                               end=avg_df_2020['day'].max(), 
                               freq='MS')  # 'MS' = Month Start

# Format the labels as month names
tick_labels = tick_locations.strftime('%b')  # 'Jan', 'Feb', etc.

# Set x-axis ticks and labels
plt.xticks(tick_locations, tick_labels, rotation=45)

plt.grid(True)
plt.legend()

# Show plot
plt.show()
```
